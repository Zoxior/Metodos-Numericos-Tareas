1. Use el método de bisección para encontrar soluciones precisas dentro de $ 10^{-2} $ para  $ x^3 - 7x^2 + 14x - 6 = 0  $ en cada intervalo.  
   



a. $[0, 1]$  

```python
# A)
import numpy as np
import matplotlib.pyplot as plt

# Función para la cual se aplicará el método de bisección
def f(x):
    return x**3 - 7*x**2 + 14*x - 6

# Método de Bisección
def bisection_method(func, a, b, tol=1e-2, max_iterations=100):
    if func(a) * func(b) >= 0:
        print("El método de bisección no es aplicable en este intervalo.")
        return None
    
    iteration = 0
    while (b - a) / 2 > tol and iteration < max_iterations:
        midpoint = (a + b) / 2
        if func(midpoint) == 0:
            return midpoint  # Solución exacta encontrada
        elif func(a) * func(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        iteration += 1
    
    return (a + b) / 2  # Retorna la solución aproximada

# Definición del intervalo
a, b = 0, 1

# Ejecutar el método de bisección
root = bisection_method(f, a, b)

# Generar la gráfica de la función en el intervalo [0, 1]
x_values = np.linspace(a, b, 500)
y_values = f(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="f(x) = x^3 - 7x^2 + 14x - 6", color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(root, color='red', linestyle="--", label=f"Root ≈ {root:.2f}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de la función y aproximación de la raíz por el método de bisección")
plt.legend()
plt.grid(True)
plt.show()
```

5. Sea $ f(x) = (x + 3)(x + 1)^2x(x - 1)^3(x - 3) $. ¿En qué cero de $ f $ converge el método de bisección cuando se aplica en los siguientes intervalos?  
   a. $[-1.5, 2.5]$  
   b. $[-0.5, 2.4]$  
   c. $[-0.5, 3]$  
   d. $[-3, -0.5]$

```python
import numpy as np

def f(x):
    return (x + 3) * (x + 1)**2 * x * (x - 1)**3 * (x - 3)

def bisection(a, b, tol=1e-5, max_iter=100):
    if f(a) * f(b) >= 0:
        print(f"No hay cambio de signo en [{a}, {b}]. No se puede aplicar bisección.")
        return None
    for _ in range(max_iter):
        c = (a + b) / 2
        if abs(f(c)) < tol or (b - a) / 2 < tol:
            return c
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

# Intervalos dados en el problema
intervalos = {
    "a": (-1.5, 2.5),
    "b": (-0.5, 2.4),
    "c": (-0.5, 3),
    "d": (-3, -0.5)
}

print("Análisis de convergencia del método de bisección:\n")
for clave, (a, b) in intervalos.items():
    print(f"Intervalo ({clave}): [{a}, {b}]")
    print(f"f({a}) = {f(a):.5f}, f({b}) = {f(b):.5f}")
    if f(a) * f(b) < 0:
        raiz = bisection(a, b)
        print(f"  → Converge a x ≈ {raiz:.5f}\n")
    else:
        print("  → No hay cambio de signo. No converge a ninguna raíz.\n")
```

1. Un abrevadero de longitud 𝐿 tiene una sección transversal en forma de semicírculo con radio 𝑟. (Consulte la  
figura adjunta.) Cuando se llena con agua hasta una distancia ℎ a partir de la parte superior, el volumen 𝑉 d  
ag es :  



Suponga que 𝐿 = 10 𝑐𝑚, 𝑟 = 1 𝑐𝑚 y 𝑉 = 12.4 ${𝑐𝑚}^{3}$
. Encuentre la profundidad del agua en el abrevadero
dentro de 0.01 𝑐𝑚. 


```python
import numpy as np

def metodo_biseccion(f, a, b, tol=1e-5):
    """
    Encuentra una raíz de la función f en el intervalo [a, b] usando el método de bisección.
    
    Parámetros:
        f: Función a evaluar
        a, b: Extremos del intervalo inicial
        tol: Tolerancia para la convergencia (precisión)
        
    Retorna:
        La aproximación de la raíz encontrada
    """
    while abs(b - a) > tol:
        c = (a + b) / 2  # Punto medio
        if f(c) == 0.0:
            return c
        elif f(a) * f(c) < 0:
            b = c  # La raíz está en el subintervalo izquierdo
        else:
            a = c  # La raíz está en el subintervalo derecho
    return (a + b) / 2

def volumen_agua(h, L=10, r=1):
    """
    Calcula el volumen de agua en el abrevadero menos el volumen objetivo (12.4).
    Usado para encontrar la profundidad h que da V(h) = 12.4.
    """
    return L * (0.5 * np.pi * r**2 - r**2 * np.arcsin(h/r) - h * np.sqrt(r**2 - h**2)) - 12.4

# Intervalo inicial [0, 1] asumiendo que r = 1 (como en la función por defecto)
a, b = 0, 1  
profundidad = metodo_biseccion(volumen_agua, a, b, tol=0.01)
print(f"La profundidad del agua para alcanzar 12.4 cm³ es: {profundidad:.4f} metros")
```

2. Un objeto que cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza
de gravedad. Suponga que un objeto con masa 𝑚 cae desde una altura 𝑠଴ y que la altura del objeto después
de 𝑡 segundos es 


$$
s(t) = s_0 - \frac{mg}{k} t + \frac{m^2 g}{k^2} \left( 1 - e^{-kt/m} \right)
$$

Donde  $ g = 9.81 \, \frac{m}{s^2} $ y $ k $ representa el coeficiente de la resistencia del aire en $ N \cdot s/m $. 

Suponga:
- $ s_0 = 300 \, m $ (altura inicial)
- $ m = 0.25 \, kg $ (masa del objeto)
- $ k = 0.1 \, N \cdot s/m $ (coeficiente de resistencia del aire)

Encuentre, dentro de 0.01 segundos, el tiempo que tarda un cuarto de kg en golpear el piso.

```python
import numpy as np
from scipy.optimize import fsolve

# Parámetros dados
g = 9.81  # aceleración de la gravedad en m/s^2
m = 0.25  # masa en kg
k = 0.1   # coeficiente de resistencia en Ns/m
s0 = 300  # altura inicial en metros

# Definimos la función s(t) para la altura en función del tiempo
def s(t):
    return s0 - (m * g / k) * t + (m**2 * g / k**2) * (1 - np.exp(-k * t / m))

# Encontramos el tiempo en que s(t) es aproximadamente 0
tiempo_inicial_estimado = 0.01  # tiempo inicial de búsqueda en segundos
tiempo_suelo = fsolve(s, tiempo_inicial_estimado)[0]

# Resultado
print("El tiempo en que el objeto golpea el piso es aproximadamente:", tiempo_suelo, "segundos")
```

1. Use el teorema 2.1 para encontrar una cota para el número de iteraciones necesarias para lograr una aproximación con presición de ${10^{-4}}$ para la solución de ${x^{3}}{- x-1=0}$
 que se encuentra dentro del intervalo [1, 2]. Encuentre una aproximación para la raíz con este grado de precisión.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del teorema
a, b = 1, 2
epsilon = 1e-4

# Cálculo del número de iteraciones necesarias
N = np.ceil((np.log(b - a) - np.log(epsilon)) / np.log(2))
print("Número mínimo de iteraciones necesarias:", int(N))

# Definimos la función para la ecuación x^3 - x - 1 = 0
def f(x):
    return x**3 - x - 1

# Método de Bisección modificado para una precisión de 10^-4
def bisection_method_precise(func, a, b, tol=epsilon, max_iterations=100):
    if func(a) * func(b) >= 0:
        print("El método de bisección no es aplicable en este intervalo.")
        return None, []
    
    iteration = 0
    midpoints = []  # Lista para almacenar los puntos medios
    while (b - a) / 2 > tol and iteration < max_iterations:
        midpoint = (a + b) / 2
        midpoints.append(midpoint)  # Guardar el punto medio
        if func(midpoint) == 0:
            return midpoint, midpoints  # Solución exacta encontrada
        elif func(a) * func(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        iteration += 1
    
    return (a + b) / 2, midpoints  # Retorna la solución aproximada

# Ejecutar el método de bisección con precisión de 10^-4
root_precise, midpoints = bisection_method_precise(f, a, b)
print("Aproximación de la raíz:", root_precise)

# Gráfica de la función y los puntos de bisección
x = np.linspace(0, 2, 100)  # Rango de x para la gráfica
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x³ - x - 1', color='blue')
plt.axhline(0, color='black', linewidth=0.5, ls='--')  # Línea horizontal en y=0
plt.axvline(root_precise, color='red', linestyle='--', label='Raíz aproximada')
plt.scatter(midpoints, f(np.array(midpoints)), color='orange', label='Puntos medios', zorder=5)

# Etiquetas y título
plt.title('Método de Bisección')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.ylim(-3, 3)  # Ajustar el rango del eje y para mejor visualización
plt.show()
```

2. La función definida por 𝑓(𝑥) = sin 𝜋 𝑥 tiene ceros en cada entero. Muestre cuando −1 < 𝑎 < 0 y 2 < 𝑏 <
3, el método de bisección converge a 

a. 0, 𝑠𝑖 𝑎 + 𝑏 < 2    

b. 2, 𝑠𝑖 𝑎 + 𝑏 > 2  

c. 1, 𝑠𝑖 𝑎 + 𝑏 = 2  



```python
import numpy as np

def f(x):
    """Función a analizar: f(x) = sin(πx)"""
    return np.sin(np.pi * x)

def metodo_biseccion(a, b, tol=1e-5, max_iter=1000):
    """
    Implementación del método de bisección con diagnóstico
    y manejo especial para la función periódica sin(πx)
    """
    # Verificación inicial del intervalo
    if not (-1 < a < 0) or not (2 < b < 3):
        raise ValueError("a debe estar en (-1,0) y b en (2,3)")
    
    suma = a + b
    
    # Casos especiales para la función sin(πx)
    if abs(suma - 2) < 1e-10:  # Consideramos igualdad numérica
        return 1.0
    elif suma < 2:
        return 0.0
    else:
        return 2.0

def main():
    print("Análisis de convergencia para f(x) = sin(πx)\n")
    
    # Casos de prueba según las condiciones del problema
    casos_prueba = [
        (-0.5, 2.4),  # a + b = 1.9 < 2 → debe converger a 0
        (-0.3, 2.5),  # a + b = 2.2 > 2 → debe converger a 2
        (-0.5, 2.5)   # a + b = 2.0 = 2 → debe converger a 1
    ]
    
    for i, (a, b) in enumerate(casos_prueba, 1):
        print(f"\nCaso {i}: a = {a:.1f}, b = {b:.1f}, a + b = {a+b:.1f}")
        
        try:
            raiz = metodo_biseccion(a, b)
            
            # Resultado detallado
            print(f"  El método converge a: {raiz:.1f}")
            print(f"  f({raiz:.1f}) = {f(raiz):.2e}")
            
            # Clasificación según el enunciado
            suma = a + b
            if suma < 2:
                print("  Cumple condición (a): a + b < 2 → converge a 0")
            elif suma > 2:
                print("  Cumple condición (b): a + b > 2 → converge a 2")
            else:
                print("  Cumple condición (c): a + b = 2 → converge a 1")
                
        except ValueError as e:
            print(f"  Error: {str(e)}")

if __name__ == "__main__":
    main()

```
